```python
# üì¶ Import Libraries
import os, zipfile
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import glob

# üìÅ Upload & Extract Good Fruits Dataset
print("Upload GOOD fruits dataset (.zip)")
good_data = files.upload()
for file in good_data:
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall("dataset/good")

# üìÅ Upload & Extract Bad Fruits Dataset
print("Upload BAD fruits dataset (.zip)")
bad_data = files.upload()
for file in bad_data:
    with zipfile.ZipFile(file, 'r') as zip_ref:
        zip_ref.extractall("dataset/bad")

# ‚úÖ Organize into train folders
os.makedirs("train_data/good", exist_ok=True)
os.makedirs("train_data/bad", exist_ok=True)

!mv dataset/good/*/* train_data/good/
!mv dataset/bad/*/* train_data/bad/

# ‚öôÔ∏è Image Preprocessing
IMG_SIZE = (128, 128)
BATCH_SIZE = 32

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_gen = datagen.flow_from_directory(
    "train_data",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

val_gen = datagen.flow_from_directory(
    "train_data",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)

# üß† Build CNN Model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(train_gen, validation_data=val_gen, epochs=5)

# üì∏ Upload Test Image
print("Upload a test fruit image or a zip file containing test images...")
test_upload = files.upload()
test_dir = "test_images"
os.makedirs(test_dir, exist_ok=True)

for file_name, content in test_upload.items():
    if file_name.endswith(".zip"):
        with open(file_name, 'wb') as f:
            f.write(content)
        with zipfile.ZipFile(file_name, 'r') as zip_ref:
            zip_ref.extractall(test_dir)
        os.remove(file_name) # Remove the zip file after extraction
    else:
        with open(os.path.join(test_dir, file_name), 'wb') as f:
            f.write(content)

# Process images in the test_dir
test_image_paths = glob.glob(os.path.join(test_dir, "**/*"), recursive=True)
test_image_paths = [f for f in test_image_paths if os.path.isfile(f) and f.lower().endswith(('.png', '.jpg', '.jpeg'))]

if not test_image_paths:
    print("No valid image files found in the uploaded data.")
else:
    for image_path in test_image_paths:
        try:
            img = Image.open(image_path).resize(IMG_SIZE)
            img_array = np.array(img) / 255.0
            img_array = np.expand_dims(img_array, axis=0)

            prediction = model.predict(img_array)[0][0]
            classification = "Spoiled" if prediction > 0.5 else "Good"
            spoilage_pct = round(prediction * 100, 2)
            edible_pct = round(100 - spoilage_pct, 2)

            print(f"\nProcessing: {os.path.basename(image_path)}")
            print(f"üçé Prediction: {classification}")
            print(f"üí• Spoilage Percentage: {spoilage_pct}%")
            print(f"üçΩÔ∏è Estimated Edible Part: {edible_pct}%")

            # üî¨ Bacteria Estimation
            def guess_bacteria(spoilage):
                if spoilage < 30:
                    return ["Low Risk"]
                elif spoilage < 60:
                    return ["Penicillium", "Rhizopus"]
                else:
                    return ["Aspergillus", "Botrytis", "Erwinia", "Rhizopus"]

            bacteria = guess_bacteria(spoilage_pct)
            print(f"ü¶† Likely Bacteria: {', '.join(bacteria)}")

            # üìä Spoilage Progression Chart
            days = list(range(1, 8))
            progression = [spoilage_pct * (i/7) for i in days]

            plt.plot(days, progression, color='red', marker='o')
            plt.title(f"üìà Spoilage Progression Over 7 Days for {os.path.basename(image_path)}")
            plt.xlabel("Days")
            plt.ylabel("Spoilage %")
            plt.grid(True)
            plt.show()

        except Exception as e:
            print(f"Could not process image {os.path.basename(image_path)}: {e}")
```
